# Task 1: Iris Flower Classification

### About Data
Overview
The Iris dataset is a well-known dataset. It contains 150 observations of iris flowers, with each observation consisting of four features and a target label. 

The dataset contains the following features (all measured in centimeters):

Sepal Length: The length of the sepal.
Sepal Width: The width of the sepal.
Petal Length: The length of the petal.
Petal Width: The width of the petal.

```python
# Encode the data
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
df.head()
```

```python
# extract independent and dependent variables 
x = df.drop(columns='Species')
y = df['Species']

# train test split
x_train,  x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# printing the shapes 
print("X_train: ",x_train.shape)
print("X_test: ",x_test.shape)
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)
```

### Model selection

```python
models = [
    ('LogisticRegression', LogisticRegression()),
    ('SVC', SVC()),
    ('DecisionTreeClassifier', DecisionTreeClassifier()),
    ('RandomForestClassifier', RandomForestClassifier()),
    ('AdaBoostClassifier', AdaBoostClassifier()),
    ('GradientBoostingClassifier', GradientBoostingClassifier()),
    ('MultinomialNB', MultinomialNB()),
    ('BernoulliNB', BernoulliNB())
]

best_model = None
best_accuracy = 0
# Iterate over the models and evaluate their performance
for name, model in models:
    # Create a pipeline for each model
    pipeline = Pipeline([
        ('model', model)
    ])
    
    pipeline.fit(x_train, y_train)
    
    # Make predictions on the test data
    y_pred = pipeline.predict(x_test)
    
    # Calculate accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    # Print the performance metrics
    print("Model:", name)
    print("Test Accuracy:", accuracy)
    print()
    
    # Check if the current model has the best accuracy
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = pipeline

import pickle
pickle.dump(best_model, open('iris_classification.pkl', 'wb'))
print("-----------------------------------------------------------")
print(f"Best model: {best_model} with accuracy score: {best_accuracy}")
```
